```python
import pandas as pd
import time
import matplotlib.pyplot as plt
```

```python
city_data_df = pd.read_csv("cities.csv")
```

```python
# Extract relevant fields from the DataFrame for plotting.
lats = city_data_df["Lat"]
max_temps = city_data_df["Max Temp"]
humidity = city_data_df["Humidity"]
cloudiness = city_data_df["Cloudiness"]
wind_speed = city_data_df["Wind Speed"]
```

```python

# Get today's date in seconds.
# today = time.time()
today = time.strftime("%x")
today
```

```python
plt.scatter(lats, max_temps, edgecolor = "black", 
           marker = "o", label = "cities", alpha = 0.8)
plt.grid()
plt.ylabel("Max Temp (F)")
plt.xlabel("Latitude")
plt.title("City Latitude vs Max Temperature "+today)
```

```python
plt.scatter(lats, humidity, edgecolor = "black", 
           marker = "o", label = "cities", alpha = 0.8)
plt.grid()
plt.ylabel("Humidity ")
plt.xlabel("Latitude")
plt.title("City Latitude vs Humidity "+today)
```

```python
# Build the scatter plots for latitude vs. cloudiness.
plt.scatter(lats,
            cloudiness,
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")

# Incorporate the other graph properties.
plt.title(f"City Latitude vs. Cloudiness (%) "+ time.strftime("%x"))
plt.ylabel("Cloudiness (%)")
plt.xlabel("Latitude")
plt.grid(True)
```

```python

plt.scatter(lats,
            wind_speed,
            edgecolor="black", linewidths=1, marker="o",
            alpha=0.8, label="Cities")

# Incorporate the other graph properties.
plt.title(f"City Latitude vs. Wind Speed "+ time.strftime("%x"))
plt.ylabel("Wind Speed (mph)")
plt.xlabel("Latitude")
plt.grid(True)
```
